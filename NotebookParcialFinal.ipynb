Miro cuales son los tipos de dato de cada una de las columnas y otra información adicional para ver que aspectos llaman la atención de estos datos e identificar que se puede hacer con ellos.  
Por ejemplo veo que la variable "Cabin" tiene demasiados nulos y creo que es una información no tan necesaria para el análisis que se me ocurre.  
La columna "Survived" está completa y tiene solo 2 opciones: 0 y 1. Con esto puedo concluir que es una variable binaria que me dice si alguien sobrevivió o no.  
Para mi análisis puedo identificar algunas columnas de interes, como "Sex", "Age", "Embarked".  
Algunas variables como "Name", "PassengerId" ó "Ticket" no me parecen muy infliyentes en el análisis que haré.  

```python
df = df.drop(["PassengerId", "Name", "Ticket", "Cabin"], axis = 1)
```

Decido eliminar las columnas: "PassengerId", "Name", "ticket" y "Cabin".  
Porque como ya mencioné algunas de ellas no las veo útiles en el enfoque de mi análisis y porque "cabin" no tiene suficiente información para  
reemplazar o completar la cantidad de información faltante.  
Esto me parece importante para evitar ruido en mi dataset y poder trabajar más fácil y cómodo con él.

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

En la información adicional de arriba se evidencia que hay columnas como "Age" y "Embarked" que contienen nulos y por esto es necesario completar la información ya que no son tantos los valores faltantes como para eliminar la columna o filas completas y así mantener la distribución de los datos lo mejor posible.  
El total de datos son 891 filas y los faltantes en cada una no superan el 20% de la inforamción.  
En edad la media es una buena manera de centralizar los datos y en embarque entendiendo que es el puerto donde se embarca reemplazar por el valor que más se repite.

```python
df = pd.get_dummies(df, columns = ["Sex", "Embarked"], drop_first = True)
```

```python
plt.figure(figsize = (8, 5))
sns.boxplot(x = "Survived", y = "Fare", data = df)
plt.show()
```

En este gráfico podemos observar la separación de datos del eje Y en si sobrevivieron o no (0 = no sobrevive, 1 = sí sobrevive) y "Fare" que es la tarifa pagada por las personas, nos sirve el gráfico de caja para identificar los diferentes cuartiles de la distribución y ver valores atípicos.  
Se evidencia que a mayor fue la tarifa pagada por la persona, mayor es el número de sobrevivientes, esto nos puede decir que hay una relación directa entre el dinero y la posición social de una persona frente a su probabilidad de sobrevivir en una catástrofe.  
Por esto identifico que la columna de tarifa es importante para entrenar mi modelo.

```python
plt.figure(figsize = (8, 5))
sns.regplot(x = "Age", y = "Survived", data = df)
plt.show()
```

Por otro lado, tenemos este gráfico lineal enfrentando si la persona sobrevive o no y su edad.  
Cómo era de esperar, se muestra una inclinación negativa, a mayor edad menos sobreviven y viceversa.  
Nos muestra la relación directa entre una variable continua "edad" y nuestra variable objetivo "Survived".  
Con esto entendemos que hay una relacióm lineal inversa entre nuestras variables.  
podemos concluir que la variable "Age" es clave para nuestro modelo predictivo.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

#Definimos la variable independiente X y la variable dependiente que queremos predecir Y.
X = df.drop("Survived", axis = 1)
y = df["Survived"]

#hacemos división de los datos en 70% entrenamiento y 30% parq pruebas.
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size = 0.3, random_state = 0)

#Ahora creamos y entrenamos nuestro modelo.
lm = LinearRegression()
lm.fit(X_train, y_train)

#hacemos nuestra primera predicción.
y_prediccion = lm.predict(X_test)

#Miramos si el modelo funciona con una evaluación y utilizando los valores aprendidos en clase para este tipo de modelo.
mse = mean_squared_error(y_test, y_prediccion)
r2 = r2_score(y_test, y_prediccion)

#Y mostramos los resultados de estos datos para analizar.
print("Mean Squared Error (MSE): ", mse)
print("R squared (R2): ", r2)
```

Las variables elegidas como predictoras (todas las demás) para el desarrollo de este modelo se justifican en constrir un perfil completo del pasajero para su posterior análisis: "Age", "Fare", "Pclass", "Sex"...  
Con esto se pudo identificar variables de peso en la probabilidad de supervivencia de la persona.  
La elección del modelo se justifica en el contenido de la clase, su clara explicación y funcionamiento para evaluar problemas con diferentes características al mismo tiempo, que es justo para lo que sirve el modelo de Regresión lineal múltiple.  

Por el lado de las métricas del MSE y R2, entendemos que el mse nos dice el promedio de los errores al cuadrado entre el valor real de supervivencia 0 o 1 y el valor continuo que predice nuestra ecuación matemática, un mse bajo indica que la distancia entre los datos reales y nuestra línea de regresión es mínima.
El R2 por su parte, indica que porcentaje de la varianza total en la supervivencia es explicada por nuestras variables independientes: "Age", "Fare", "Pclass", "Sex"...  

Yo creo que el modelo responde bien y coherentemente sobre todo, al aplicar Regresión Lineal Múltiple a una variable binaria 0 y 1 estamos creando un modelo lneal de probabilidad, es normal que el R2 no sea cercano a 1 ya que tenemos datos binarios polarizados y la predicción se hace por medio de una linea continua, luego que el mse nos de bajo es una buena señal de que el modelo responde bien a la tendencia lógica, por ejemplo de que ser niño ó tener recursos económicos aumentan la probabilidad de sobrevivir.
