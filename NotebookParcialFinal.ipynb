Miro cuales son los tipos de dato de cada una de las columnas y otra información adicional para ver que aspectos llaman la atención de estos datos e identificar que se puede hacer con ellos.  
Por ejemplo veo que la variable "Cabin" tiene demasiados nulos y creo que es una información no tan necesaria para el análisis que se me ocurre.  
La columna "Survived" está completa y tiene solo 2 opciones: 0 y 1. Con esto puedo concluir que es una variable binaria que me dice si alguien sobrevivió o no.  
Para mi análisis puedo identificar algunas columnas de interes, como "Sex", "Age", "Embarked".  
Algunas variables como "Name", "PassengerId" ó "Ticket" no me parecen muy infliyentes en el análisis que haré.  

```python
df = df.drop(["PassengerId", "Name", "Ticket", "Cabin"], axis = 1)
```

Decido eliminar las columnas: "PassengerId", "Name", "ticket" y "Cabin".  
Porque como ya mencioné algunas de ellas no las veo útiles en el enfoque de mi análisis y porque "cabin" no tiene suficiente información para  
reemplazar o completar la cantidad de información faltante.  
Esto me parece importante para evitar ruido en mi dataset y poder trabajar más fácil y cómodo con él.

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

En la información adicional de arriba se evidencia que hay columnas como "Age" y "Embarked" que contienen nulos y por esto es necesario completar la información ya que no son tantos los valores faltantes como para eliminar la columna o filas completas y así mantener la distribución de los datos lo mejor posible.  
El total de datos son 891 filas y los faltantes en cada una no superan el 20% de la inforamción.  
En edad la media es una buena manera de centralizar los datos y en embarque entendiendo que es el puerto donde se embarca reemplazar por el valor que más se repite.

```python
df = pd.get_dummies(df, columns = ["Sex", "Embarked"], drop_first = True)
```
